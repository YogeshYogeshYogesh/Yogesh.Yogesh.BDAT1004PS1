# What data type is each of the following?
# 5 
# 5.0
# 5 > 1
# '5'
# 5 * 2
# '5' * 2
# '5' + '2'
# 5 / 2
# 5 % 2 {5, 2, 1}
# 5 == 3
# Pi (the number)

```python
using System;

namespace ConsolApp1
{
  class Program
  {
    static void Main(string[] args)
    {
      /*we use Math library only for PI
             *  var is used if data type is unknown
             * Type Class if used to find datatype*/
            var a = 5;
            Type t = a.GetType();
            Console.WriteLine("Data Type of 5 is  " + t.ToString());

            var b = 5.0;
            Type u= b.GetType();
            Console.WriteLine("Data Type of 5.0 is  " + u.ToString());

            var c = 5>1 ;
            Type v = c.GetType();
            Console.WriteLine("Data Type of '5' is  " + v.ToString());

            var d = '5';
            Type w = d.GetType();
            Console.WriteLine("Data Type of '5' is  " + w.ToString());

            var e = 5*2;
            Type x = e.GetType();
            Console.WriteLine("Data Type of 5*2 is  " +x.ToString());

            var f = '5'*2 ;
            Type y = f.GetType();
            Console.WriteLine("Data Type of '5'*2 is  " + y.ToString());

            var g= '5'+'2';
            Type z = g.GetType();
            Console.WriteLine("Data Type of '5'+'2' is  " + z.ToString());

            var h = 5 /2;
            Type zz = h.GetType();
            Console.WriteLine("Data Type of 5%2 is  " + zz.ToString());

            var i = 5%2;
            Type zy =i.GetType();
            Console.WriteLine("Data Type of 5%2 is  " + zy.ToString());

            var j = new []{ 5, 2, 1 };
            Type zx = j.GetType();
            Console.WriteLine("Data Type of { 5, 2, 1 } is  " + zx.ToString());

            var k = 5==3;
            Type tu = k.GetType();
            Console.WriteLine("Data Type of 5==3 is  " + tu.ToString());

        
        /*we use Math library only for PI
             *  var is used if data type is unknown
             * Type Class if used to find datatype*/
        
            var l = Math.PI;
            Type tv = l.GetType();
            Console.WriteLine("Data Type of PI is  " + tv.ToString());


        }   
    }
  }
```
